# 01 — Data Loading and Cleaning
This notebook demonstrates how to load the e-commerce CSV files into Pandas, inspect the data, check data types, identify missing values, and perform basic cleaning.

**Concepts demonstrated:** `pd.read_csv`, `head`, `info`, `isnull`, `drop_duplicates`, `to_datetime`

```
import pandas as pd
from pathlib import Path
```

## 1. Define file paths
Adjust the `RAW` path if your repo structure is different.

```
RAW = Path('../data/raw')
orders_path = RAW / 'orders.csv'
customers_path = RAW / 'customers.csv'
order_items_path = RAW / 'order_items.csv'
products_path = RAW / 'products.csv'
```

## 2. Load the datasets

```
orders = pd.read_csv(orders_path)
customers = pd.read_csv(customers_path)
order_items = pd.read_csv(order_items_path)
products = pd.read_csv(products_path)

display(orders.head())
display(customers.head())
```

## 3. Inspect structure and missing values

```
print('Orders info:')
orders.info()

print('\nMissing values in orders:')
print(orders.isnull().sum())
```

## 4. Convert date columns and remove duplicates

```
orders_clean = orders.copy()
for col in ['order_date', 'shipped_date', 'delivered_date']:
    if col in orders_clean.columns:
        orders_clean[col] = pd.to_datetime(orders_clean[col], errors='coerce')

before = len(orders_clean)
orders_clean = orders_clean.drop_duplicates()
after = len(orders_clean)
print(f'Removed {before - after} duplicate rows')
orders_clean.head()
```

## 5. Basic profiling
This is useful for data quality review before any SQL or BI work.

```
print('Order status counts:')
print(orders_clean['order_status'].value_counts(dropna=False))

print('\nCustomers by state:')
if 'state' in customers.columns:
    print(customers['state'].value_counts().head(10))
```
