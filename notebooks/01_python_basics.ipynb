## 1️⃣ Setup + First Program
**Concepts:** `print()` function, *Strings*

## 2️⃣ Variables – Storing Business Facts
**Concepts:** Variables (`=`), numbers, strings, booleans

**Business Context:** Store metrics like total orders, spend, and active status for each customer.

```
customer_name = "Ann"
total_orders = 15
total_spent = 1280.50
is_active = True

print(customer_name, "placed", total_orders, "orders worth", total_spent)
```

## 3️⃣ Data Types + Conversion
**Concepts:** `type()`, casting (`int()`, `float()`, `str()`)

**Business Context:** Convert text data from CSVs to numeric types before aggregation.

```
revenue_text = "5000"
revenue = int(revenue_text)
print(type(revenue))
print(revenue + 250)
```

## 4️⃣ Operators – Building KPIs
**Concepts:** Arithmetic (`+ - * / // %`), rounding with `round()`

**Business Context:** Compute metrics like Average Order Value (AOV) or Fulfillment Rate.

```
revenue = 12500
orders = 250
aov = revenue / orders
print("Average Order Value (AOV):", round(aov, 2))
```

## 5️⃣ Strings – Working with Text
**Concepts:** String concatenation, f-strings, methods (`.upper()`, `.title()`, `len()`)

**Business Context:** Clean inconsistent category or region labels before grouping.

```
category = "electronics"
print(category.title())

customer = "Alice"
product = "Headphones"
price = 99.99
print(f"{customer} bought {product} for ${price}")
```

## 6️⃣ Lists – Collections of Items
**Concepts:** Lists `[ ]`, indexing, `.append()`, `for` loop iteration

**Business Context:** Use lists for storing product categories or months for reporting.

```
products = ["Laptop", "Phone", "Tablet"]
products.append("Smartwatch")
for item in products:
    print("Analyzing sales for:", item)
```

## 7️⃣ Dictionaries – Structured Records
**Concepts:** Key-value pairs `{}`, dictionary access `dict['key']`

**Business Context:** Represent an order or customer record, similar to a JSON object.

```
order = {"order_id": 101, "customer": "Ann", "amount": 250.75, "status": "delivered"}
print(order["customer"], "spent $", order["amount"])
```

## 8️⃣ Conditionals – Business Rules
**Concepts:** `if`, `elif`, `else`, comparison operators (`>`, `<`, `==`)

**Business Context:** Check thresholds like free shipping or high refund rate.

```
order_value = 180
if order_value >= 200:
    print("Free shipping")
else:
    print("Shipping fee applies")
```

## 9️⃣ Loops – Automating Repetition
**Concepts:** `for` loop, `range()`, accumulators (`+=`)

**Business Context:** Iterate over lists of orders, months, or categories to compute totals.

```
order_values = [120, 250, 180, 300]
total = 0
for value in order_values:
    total += value
print("Total revenue:", total)
```

## 🔟 Functions – Reusable Analytics Blocks
**Concepts:** `def`, parameters, `return`, scope

**Business Context:** Define reusable calculations (AOV, ROAS, Refund Rate).

```
def average_order_value(total_revenue, num_orders):
    """Compute Average Order Value (AOV)."""
    return round(total_revenue / num_orders, 2)

print(average_order_value(1280.5, 15))
```

## ⭐ Mini Project – Customer Summary Function
Combine all previous concepts into a customer summary function.

**Concepts:** Function composition, arithmetic, f-strings

**Business Context:** Generate quick summaries like a CRM card.

```
def order_summary(customer, orders, spend):
    avg = spend / orders
    print(f"{customer} placed {orders} orders totaling ${spend}.")
    print("Average Order Value:", round(avg, 2))

order_summary("Ann", 15, 1280.5)
```
